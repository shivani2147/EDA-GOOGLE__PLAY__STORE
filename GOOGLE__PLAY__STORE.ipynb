# INTRODUCTION

Each column represents a specific attribute of the apps. Here's a brief explanation of what each column represents:

- App: The name of the mobile application.
- Category: The category to which the app belongs (e.g., "Business," "Entertainment," "Education").
- Rating: The user rating of the app (typically ranges from 0 to 5).
- Reviews: The number of user reviews for the app.
- Size: The size of the app.
- Installs: The number of times the app has been downloaded/installed.
- Type: The type of the app (e.g., "Free" or "Paid").
- Price: The price of the app (if it's a paid app).
- Content Rating: The age group for which the app's content is appropriate.
- Genres: Specific genres or tags associated with the app.
- Last Updated: The date when the app was last updated.
- Current Ver: The current version of the app.
- Android Ver: The minimum Android version required to run the app.

```python
# Import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset

df = pd.read_csv("D:\Datasets\googleplaystore.csv")
df.sample(10)
```

```python
# Shape (Rows & Cols) of the dataset
df.shape
```

```python
# Remove the unwanted columns 
df.drop(['Current Ver', 'Android Ver'], axis = 1, inplace = True)
```

```python
# Check null values
df.isnull().sum()

# Here, most of the null values are in Rating column
```

```python
# Check datatypes 
df.dtypes

# Reviews, Size, Installs, Price --> will convert these into float type
```

## Rating

```python
# Fill the null values in Rating col with mean
df['Rating'].fillna(df['Rating'].mean(), inplace = True)
```

## Reviews

```python
# Check which is non-numeric in Reviews column

count = 0
for idx, data in enumerate(df['Reviews']) : 
    if not data.isnumeric() : 
        count += 1
        print(idx, '', data)
        
print("No. of Special Characters : ", count)
```

```python
# Check 3.0M
df[df['Reviews'] == '3.0M']
```

```python
# Drop 10472 row because all the values are shifted here
df.drop(10472, inplace = True)
```

```python
# Change the datatype of Reviews column from object to float
df['Reviews'] = df['Reviews'].astype('float64')
```

## Size

```python
# Remove 'M' from data of Size column
df['Size'] = df['Size'].map(lambda X : X.rstrip('M'))
```

```python
"""
1 MB = 1024 KB

To convert KB into MB : 
1024 KB = 1 MB
1 KB = 1 / 1024 MB
"""
```

```python
# Remove K in Size column, after that change the units from 'K' to 'M' divided by 1024
# Convert KB into MB

df['Size'] = df['Size'].map(lambda X : str(round((float(X.strip('k')) / 1024), 1)) if X[-1] == 'k' else X)
```

```python
# Replace 'Varies with device' with null value
df['Size'].replace('Varies with device', np.nan, inplace = True)
```

```python
# Change the datatype of Size column from object to float
df['Size'] = df['Size'].astype('float64')
```

```python
# Check null values
df.isnull().sum() 

# There came null values in this column
```

```python
# Replace null values of Size col with its mean
df['Size'].fillna(df['Size'].mean(), inplace = True)
```

### Installs

```python
# Remove '+' sign from the Installs column
df['Installs'] = df['Installs'].map(lambda X : X.rstrip('+'))
```

```python
# Change datatype of Installs column and also remove commas from the values
df['Installs'] = df['Installs'].str.replace(",", '').astype('float64')
```

### Price

```python
# Check how much non-numerics in Price column

count = 0
for idx, data in enumerate(df['Price']) : 
    if not data.isnumeric() : 
        count += 1
        
print("No. of Special Characters : ", count)
```

```python
# Remove '$' sign from Price column
df['Price'] = df['Price'].map(lambda X : X.lstrip('$'))
```

```python
# Change datatype of Price column
df['Price'] = df['Price'].astype('float64')
```

### Last Updated

```python
# Check Last Updated
df['Last Updated'].unique()
```

```python
# Change values into date format
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
# Remove all the duplicate values
df.dropna(inplace = True)
```

# Percentage of Review Sentiments (Positive, Negative & Neutral)

```python
# Check list of Reviews counts
# list(df['Reviews'].value_counts())
```

```python
# Define sentiment thresholds
positive_threshold = 130           # Reviews above this value are considered Positive
negative_threshold = 20            # Reviews above this value are considered Negative

# Classify reviews into positive, negative, or neutral based on sentiment thresholds
df['Review_Sentiment'] = df['Reviews'].apply(lambda reviews : 'Positive' if reviews >= positive_threshold 
                                            else 'Negative' if reviews <= negative_threshold else 'Neutral')

# Show pie chart on Sentiment 
plt.figure(figsize = (10, 6))
df['Review_Sentiment'].value_counts().plot(kind = 'pie', autopct = '%1.f%%', explode = [0, 0, 0.1])

# Add Title
plt.title("Percentage of Review Sentiments", fontsize = 15, fontweight = 600)

# Display the chart
plt.legend()
plt.show()
```

# Category vs Total App installs

```python
# Grouping the dataset by 'Category' and calculating the sum of 'Installs' for each category,
# then sorting the results in descending order.
category_installs = df.groupby('Category')['Installs'].sum().sort_values(ascending = False)

# Create a bar plot for Category vs. Total App Installs
plt.figure(figsize=(12, 8))
plt.bar(category_installs.index, category_installs.values, color = 'green')

# Add labels and title
plt.title('Categories Based on Total App Installs', fontsize = 20, fontweight = 600)
plt.xlabel('Category', fontsize = 15)
plt.ylabel('Total App Installs', fontsize = 15)

# Rotate category names for better visibility
plt.xticks(rotation = 90, ha ='right')

# Display the plot
plt.tight_layout()
plt.show()
```

---
- The chart has a bar for each category of app, and the height of each bar represents the number of installs for that category.
- It shows that the most popular categories of apps are those that people use for entertainment, communication, and productivity.
---

# Count of Apps in Each Category differentiated by Their Type

```python
# Group the data by both 'Category' and 'Type', and count the occurrences of each combination
category_type_counts = df.groupby(['Category', 'Type']).size().unstack()

# Create a chart to show
category_type_counts.plot(kind = 'barh', stacked = True)

# Add labels and title
plt.title('Count of Applications in Each Category by Type', fontsize = 15, fontweight = 600)
plt.ylabel('Category', fontsize = 12)
plt.xlabel('Count of Apps', fontsize = 12)

# Display the chart 
plt.show()
```

---

- The chart shows that there are more free apps than paid apps in most categories, except for weather, video players, and photography.

###### There are a few reasons why there are more paid apps in the weather, video players, and photography categories.

- Accuracy and reliability : Weather apps need to be accurate and reliable in order to be useful. This can be difficult and expensive to achieve, so developers of weather apps may charge a fee to cover the costs of development and maintenance.
- Features : Video players and photography apps often have a lot of features that can be expensive to develop. Developers of these apps may charge a fee to recoup their development costs and make a profit.
- Branding : Some developers of weather, video players, and photography apps want to create a premium brand for their products. Charging a fee for these apps can help to create a sense of exclusivity and value.

# Distribution of Content Ratings

```python
# Count the occurrences of each Content Rating
content_rating_counts = df['Content Rating'].value_counts()

plt.plot(content_rating_counts.index, content_rating_counts.values, marker = 'o', color = 'purple')

# Add labels and title
plt.title('Distribution of Content Ratings', fontsize = 15, fontweight = 600)
plt.xlabel('Content Rating', fontsize = 12)
plt.ylabel('Number of Apps', fontsize = 12)

# Rotate content ratings for better visibility
plt.xticks(rotation = 45, ha = 'right')

# Display the plot
plt.tight_layout()
plt.show()
```

---
Here's an explanation of each content rating category:

- Everyone : 
    - Apps with this content rating are suitable for all age groups. They may contain minimal or no objectionable content.

- Teen : 
    - Apps with this content rating are generally suitable for ages 13 and older. They may contain mild violence, suggestive themes, or infrequent use of strong language.

- Mature 17+ : 
    - Apps with this content rating are intended for ages 17 and older. They may contain intense violence, sexual content, or strong language.

- Everyone 10+ : 
    - Apps with this content rating are suitable for ages 10 and older. They may contain more cartoon, fantasy, or mild violence compared to the "Everyone" category.

- Adults only 18+ : 
    - Apps with this content rating are designed for adults only, typically ages 18 and older. They may contain explicit sexual content, intense violence, or strong language.

- Unrated : 
    - This indicates that the app does not have a content rating assigned. It might be new or not reviewed yet.

# Category vs. Genres
